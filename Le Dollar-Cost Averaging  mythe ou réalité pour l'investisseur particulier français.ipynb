```python
import yfinance as yf
import pandas as pd
```

# En marché haussier : le MSCI World (2010-2019) 

```python
# Télécharger les données mensuelles CW8 
msci = yf.download("CW8.PA", start="2010-01-01", end="2019-12-31", interval="1mo")

# Paramètres
capital = 60000
per_month = 500
```

```python
# Prix
prix_debut = msci["Close"].iloc[0]
prix_fin = msci["Close"].iloc[-1]  
```

```python
# LUMP SUM
ls_parts = capital / prix_debut
ls_capital_final = ls_parts * prix_fin
ls_gain = ls_capital_final - capital
ls_rendement = (ls_gain / capital) * 100
```

```python
# DCA
msci["DCA_parts"] = per_month / msci["Close"]
dca_total_parts = msci["DCA_parts"].sum()
dca_total_investi = per_month * len(msci)
dca_cout_moyen = dca_total_investi / dca_total_parts
dca_capital_final = dca_total_parts * prix_fin
dca_gain = dca_capital_final - dca_total_investi
dca_rendement = (dca_gain / dca_total_investi) * 100
```

```python
# RÉSULTATS
print("  BACKTEST CW8 — DCA vs LUMP SUM (2010-2019)")

print(f"\n  Performance CW8 : de {float(prix_debut):.2f}€ à {float(prix_fin):.2f}€ (+{float((prix_fin/prix_debut - 1)*100):.1f}%)")

print(f"\n  {'':25} {'DCA':>15} {'Lump Sum':>15}")
print(f"  {'-'*55}")
print(f"  {'Capital investi':25} {float(dca_total_investi):>12,.0f} € {float(capital):>12,.0f} €")
print(f"  {'Coût moyen entrée':25} {float(dca_cout_moyen):>12.2f} € {float(prix_debut):>12.2f} €")
print(f"  {'Capital final':25} {float(dca_capital_final):>12,.0f} € {float(ls_capital_final):>12,.0f} €")
print(f"  {'Gain':25} {float(dca_gain):>11,.0f} €  {float(ls_gain):>11,.0f} €")
print(f"  {'Rendement':25} {float(dca_rendement):>11.1f} %  {float(ls_rendement):>11.1f} %")

print(f"\n  Écart : Lump Sum gagne de {float(ls_capital_final - dca_capital_final):,.0f} €")
print(f"  Nombre de versements DCA : {len(msci)}")
print("=" * 55)
```

# En marché de crise : le MSCI World (2019-2023) 

```python
# Télécharger les données mensuelles CW8
msci = yf.download("CW8.PA", start="2019-01-01", end="2023-12-31", interval="1mo")
```

```python
# Paramètres
nb_mois = len(msci)
per_month = 500
capital = per_month * nb_mois 
```

```python
# Prix
prix_debut = msci["Close"].iloc[0]
prix_fin = msci["Close"].iloc[-1]
```

```python
# LUMP SUM
ls_parts = capital / prix_debut
ls_capital_final = ls_parts * prix_fin
ls_gain = ls_capital_final - capital
ls_rendement = (ls_gain / capital) * 100
```

```python
# DCA
msci["DCA_parts"] = per_month / msci["Close"]
dca_total_parts = msci["DCA_parts"].sum()
dca_total_investi = per_month * nb_mois
dca_cout_moyen = dca_total_investi / dca_total_parts
dca_capital_final = dca_total_parts * prix_fin
dca_gain = dca_capital_final - dca_total_investi
dca_rendement = (dca_gain / dca_total_investi) * 100
```

```python
# RÉSULTATS
print("  BACKTEST CW8 — DCA vs LUMP SUM (2019-2023)")
 
print(f"\n  Nombre de mois : {nb_mois}")
print(f"  Capital total  : {float(capital):,.0f} €")
print(f"  Performance CW8 : {float(prix_debut):.2f}€ → {float(prix_fin):.2f}€ (+{float((prix_fin/prix_debut - 1)*100):.1f}%)")
 
print(f"\n  {'':25} {'DCA':>15} {'Lump Sum':>15}")
print(f"  {'-'*55}")
print(f"  {'Capital investi':25} {float(dca_total_investi):>12,.0f} € {float(capital):>12,.0f} €")
print(f"  {'Coût moyen entrée':25} {float(dca_cout_moyen):>12.2f} € {float(prix_debut):>12.2f} €")
print(f"  {'Capital final':25} {float(dca_capital_final):>12,.0f} € {float(ls_capital_final):>12,.0f} €")
print(f"  {'Gain':25} {float(dca_gain):>11,.0f} €  {float(ls_gain):>11,.0f} €")
print(f"  {'Rendement':25} {float(dca_rendement):>11.1f} %  {float(ls_rendement):>11.1f} %")
 
ecart = ls_capital_final - dca_capital_final
gagnant = "LUMP SUM" if float(ecart) > 0 else "DCA"
print(f"\n  Gagnant : {gagnant} (écart : {abs(float(ecart)):,.0f} €)")
print("=" * 55)
```
